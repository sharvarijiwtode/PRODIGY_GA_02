```python
import torch
from diffusers import StableDiffusionPipeline

# Load the model (make sure you have accepted model terms on Hugging Face)
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16,
).to("cuda")
```

```python
# List of prompts
prompts = [
    "a fantasy landscape with glowing mushrooms",
    "a robot playing violin in a futuristic concert hall",
    "sunset over snowy mountains",
    "a cute dog in astronaut suit on the moon",
    "a magical tree with stars instead of leaves"
]

# Generate and save images
for i, prompt in enumerate(prompts):
    image = pipe(prompt).images[0]
    image_path = f"generated_image_{i+1}.png"
    image.save(image_path)
    print(f"Saved: {image_path}")
```
